```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Wczytanie danych z pliku CSV
data = pd.read_csv('simulation_agents.csv')
```

```python
def plot_strategy_vs_survival(data):
    plt.figure(figsize=(8, 6))
    survival_strategy = data.groupby(['Strategy', 'Survived']).size().unstack(fill_value=0)
    survival_strategy.plot(kind='bar', stacked=True, color=['skyblue', 'orange'])
    plt.title('Survival by Strategy')
    plt.xlabel('Strategy')
    plt.ylabel('Count')
    plt.legend(['Survived', 'Died'], title='Outcome')
    plt.tight_layout()
    plt.show()

plot_strategy_vs_survival(data)
```

```python
def plot_age_vs_survival(data):
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=data, x='Survived', y='Age', palette=['skyblue', 'orange'])
    plt.title('Age vs Survival')
    plt.xlabel('Survived (0 = No, 1 = Yes)')
    plt.ylabel('Age')
    plt.tight_layout()
    plt.show()
    
plot_age_vs_survival(data)
```

```python
def plot_avg_escape_time_by_strategy(data):
    plt.figure(figsize=(8, 6))
    avg_escape_time = data.groupby('Strategy')['Escape_Time'].mean()
    avg_escape_time.plot(kind='bar', color='skyblue')
    plt.title('Average Escape Time by Strategy')
    plt.xlabel('Strategy')
    plt.ylabel('Average Escape Time')
    plt.tight_layout()
    plt.show()

plot_avg_escape_time_by_strategy(data)
```

```python
def plot_survivors_with_relatives(data):
    plt.figure(figsize=(8, 6))
    relatives_survival = data[data['Survived'] == 1]['Has_Relatives'].value_counts()
    relatives_survival.plot(kind='bar', color=['gray', 'purple'])
    plt.title('Survivors with Relatives')
    plt.xlabel('Has Relatives')
    plt.ylabel('Count')
    plt.xticks(ticks=[0, 1], labels=['No', 'Yes'], rotation=0)
    plt.tight_layout()
    plt.show()
plot_survivors_with_relatives(data)
```

```python
def plot_risk_level_vs_survival(data):
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=data, x='Survived', y='Risk_Level', palette=['red', 'green'])
    plt.title('Risk Level vs Survival')
    plt.xlabel('Survived (0 = No, 1 = Yes)')
    plt.ylabel('Risk Level')
    plt.tight_layout()
    plt.show()
    
plot_risk_level_vs_survival(data)
```

```python
def plot_avg_health_by_strategy(data):
    plt.figure(figsize=(8, 6))
    avg_health = data.groupby('Strategy')['Health_End'].mean()
    avg_health.plot(kind='bar', color='orange', )
    plt.title('Average Health by Strategy')
    plt.xlabel('Strategy')
    plt.ylabel('Average Health')
    plt.tight_layout()
    plt.show()

plot_avg_health_by_strategy(data)
```

```python
def plot_evacuations_histogram(data):
    plt.figure(figsize=(8, 6))
    plt.hist(data['Escape_Time'], bins=15, color='skyblue', edgecolor='black')
    plt.title('Histogram of Evacuation Times')
    plt.xlabel('Escape Time')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

plot_evacuations_histogram(data)
```

```python
def plot_cumulative_rescued_agents(data):
    data_sorted = data.sort_values('Escape_Time')
    data_sorted['Cumulative_Rescued'] = data_sorted['Survived'].cumsum()
    plt.figure(figsize=(8, 6))
    plt.plot(data_sorted['Escape_Time'], data_sorted['Cumulative_Rescued'], marker='o', color='green')
    plt.title('Cumulative Number of Rescued Agents Over Time')
    plt.xlabel('Escape Time')
    plt.ylabel('Cumulative Rescued Agents')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_cumulative_rescued_agents(data)

```
